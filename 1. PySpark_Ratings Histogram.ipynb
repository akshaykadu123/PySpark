```
#Aim is to see the numbers of ratings received for each ratings level and plotting a histogram
#Here we use "countByValue()" function in PySpark.
```

```
from pyspark import SparkConf, SparkContext
import collections  #Used to sort a dictionary
```

```
conf = SparkConf().setMaster("local").setAppName("RatingsHistogram")  #Running on local machine and setting app name
sc = SparkContext.getOrCreate(conf = conf)  #Create a spark context object
```

```
lines = sc.textFile("/FileStore/tables/u.data")  #Create a RDD by reading the original complete text data file from Distributed File system in Databricks
#This reads each line
#To collect all the lines
llist = lines.collect()  
for line in llist:
  print(line)   #To see the complete dataset (all the lines)
```

```
lines.count()  #Count the number of elements (lines) in the RDD.
```

```
groupBy = lines.groupBy(lambda x: x.split()[0]).count()  #Grouping by MovieID by taking the count
groupBy
```

```
# If it is a table/dataframe, we use "table.show(5)" command to show the top 5 rows.
# For a dataframe, we use "display(dataframe)" command to see the dataframe.
# For a dataframe, we use "file.read_csv"to read the file.
```

```
#We want to extract the 3rd column from this text file (Note: This is not a table, so we cannot subset/filter a column)
#Therefore we use 'map' function and split function to split each line.
ratings = lines.map(lambda x: x.split()[2]) #Creating a rdd called 'ratings' which includes only the ratings part
#where 'x' is each line

rating_list = ratings.collect()  #To see the 'ratings' column
for rating in rating_list:
  print(rating)
```

```
#See first 10 ratings from the rating_list
rating_list
```

```
# Perform the action we want
# Here we want the frequency or count of each rating

result = ratings.countByValue()

type(result)

result
```

```
#Print the result
for i in result:
  print (i, result[i])
```

```
#Another method to print using d.items()
for key, value in result.items():    #For Python 3 --> dictionary.items()
  print (key, value)
  
```

```
#Sorting by rating key:
sorted_result1 = collections.OrderedDict(sorted(result.items()))  #This is a sorted DICTIONARY
for key, value in sorted_result1.items():
  print (key, value)    #Because it's a dictionary
```

```
#Another method to sort by rating key:
sorted_result2 = sorted(result.items())   #This always returns a sorted LIST and not dictionary
for i in sorted_result2:
  print (i[0], i[1])   #Because it's a list   #You can also use key, value here
#OR
sorted_result2
```
