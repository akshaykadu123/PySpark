```
from pyspark import SparkConf, SparkContext

conf = SparkConf().setMaster("local").setAppName("CustomerAmount")
sc = SparkContext.getOrCreate(conf=conf)
```

```
data = sc.textFile("/FileStore/tables/customer_orders.csv")
for i in data.collect()[0:5]:
  print(i)
```

```
#The above data is [1] CustomerID [2] ItemID [3] AmountSpent
#Therefore we only need CustomerID and AmountSpent

IdAmount = data.map(lambda x: x.split(",")).map(lambda x: (int(x[0]), float(x[2])))  #Creating an RDD of required data only

for i in IdAmount.collect()[0:5]:
  print (i)
```

```
summation = IdAmount.reduceByKey(lambda x,y: round(x+y, 2))   #As we are using 'reduceByKey()' we only need to consider Values. Ignore key.
#'x' is VALUE in the first row, 'y' is VALUE in the second row.

results = summation.sortByKey()

for i in results.collect():
  print(i)
```
